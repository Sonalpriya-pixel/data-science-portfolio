```python
import seaborn as sns

df = sns.load_dataset("titanic")
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
sns.countplot(x="survived", data=df)
plt.title("total survived count")
```

```python
sns.countplot(x="pclass" , hue="survived" , data=df)
plt.title("survived count by pclass")
```

```python
sns.countplot(x="sex" , hue="survived" , data=df)
plt.title("survived count by sex")
```

```python
sns.histplot(x="age" , hue="survived" , data=df)
plt.title("survived count by Age")
```

```python
sns.histplot(x="fare" , hue="survived" , data=df)
plt.title("survived count by fare")
```

```python
sns.boxplot(x="pclass" , y="age" , data=df)
plt.title("age by pclass")
```
