```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sample_student_performance.csv")
```

```python
df
```

```python
df.describe()
```

```python
sns.histplot(df['G3'],kde = True)
plt.title('Distribution of Final Grade')
plt.xlabel('Final Grade for G3')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(data = df[['G1', 'G2', 'G3']])
plt.title("Boxplot for grades")
plt.show()
```

```python
print(df.columns)
```

```python
df['gender_number'] = df['gender'].map({'male': 1, 'female': 0})
df.drop('gender', axis=1, inplace=True)
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
grouped = df.groupby(['gender_number', 'study_time'])['G3'].mean().unstack()
grouped.plot(kind='bar')
plt.title('Average Final Grade (G3) by Gender and Study Time')
plt.xlabel('Average G3')
plt.show()
```

```python
summary = df.describe()
summary.to_csv('summary_statistics.csv')
print("Summary statistics saved to summary_statistics.csv")
```
